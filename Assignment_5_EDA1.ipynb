```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading the dataset**

```python
df=pd.read_csv('/content/Cardiotocographic.csv')
df
```

**Finding Missing Value**

```python
df.isnull().sum()
```

**Drop rows with missing values**

```python
df.dropna(inplace=True)
```

 Statistical Summary

```python
df.describe()
```

**Data Visualization**

```python
# Histograms for distribution

df.hist(figsize=(15, 10), bins=20)
plt.suptitle("Histograms of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
# Boxplots to check outliers
plt.figure(figsize=(15,8))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Boxplots of Features")
plt.show()

```

```python
#Correlation Heatmap

plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Pair Plot
sns.pairplot(df.sample(100))  # Use sample to speed up
plt.suptitle("Pairplot of Sample Data", y=1.02)
plt.show()
```

**Pattern Recognition **

```python
correlations = df.corr()["NSP"].sort_values(ascending=False)
print("Correlation of features with NSP:\n", correlations)
```

CONCLUSION:

Through a comprehensive exploratory data analysis (EDA) of the fetal cardiographic dataset, several key insights and patterns were discovered.

1)No major data type issues, and missing values were handled.

2)Key features like ASTV, MSTV, and LB showed strong correlations with fetal state (NSP), making them important indicators.

3)Distributions revealed outliers and skewness in variables like DP and DS.

4)Heatmaps and scatter plots highlighted meaningful relationships and class separability.
